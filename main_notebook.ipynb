```python
import numpy as np
import random

def convert_color_matrix(color_matrix):
    """Convert a matrix of letters into a numerical matrix with assigned random colors."""
    
    # Define a set of basic but different colors
    basilar_colors = [
        (255, 0, 0),   # Red
        (0, 255, 0),   # Green
        (0, 0, 255),   # Blue
        (255, 255, 0), # Yellow
        (255, 165, 0), # Orange
        (128, 0, 128), # Purple
        (0, 255, 255), # Cyan
        (255, 192, 203), # Pink
        (139, 69, 19), # Brown
        (0, 128, 128)  # Teal
    ]
    
    unique_letters = np.unique(color_matrix)  # Extract unique letters
    random.shuffle(basilar_colors)  # Shuffle colors to assign randomly

    # Create a mapping from letters to integers and colors
    letter_to_int = {letter: i for i, letter in enumerate(unique_letters)}
    color_dict = {i: basilar_colors[i % len(basilar_colors)] for i in range(len(unique_letters))}

    # Convert matrix from letters to indices
    indexed_matrix = np.vectorize(letter_to_int.get)(color_matrix)

    # Reshape to add extra dimension: [[index, 0]]
    indexed_matrix = np.array([[[idx, 0] for idx in row] for row in indexed_matrix])

    return indexed_matrix, color_dict
```
